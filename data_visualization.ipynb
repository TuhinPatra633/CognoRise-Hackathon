```python
value_counts = data["test preparation "].value_counts()
ax = value_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding count at the top of each bar
for i in ax.containers:
    ax.bar_label(i)

# Adding title and labels
plt.title('Test Preparation Distribution In Data')
plt.xlabel('Test Preparation')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Displaying the plot
plt.show()
```

```python
value_counts = data["gender"].value_counts()
ax = value_counts.plot(kind='pie', colors=['blue', 'orange'], autopct='%1.1f%%', startangle=90, counterclock=False)

# Adding title
plt.title('Gender Distribution in Data')
plt.ylabel('') 

# Displaying the plot
plt.show()
```

```python
value_counts = data["parental level of education"].value_counts()
plt.figure(figsize=(12, 8)) 
ax = value_counts.plot(kind='bar', color='lightgreen', edgecolor='black')

# Adding count at the top of each bar
for i in ax.containers:
    ax.bar_label(i)

# Adding title and labels

plt.title('Parental level of education In Data')
plt.xlabel('Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

```python
value_counts = data["lunch"].value_counts()
plt.figure(figsize=(10, 6)) 
ax = value_counts.plot(kind='bar', color="lightblue", edgecolor='black')

# Adding count at the top of each bar
for i in ax.containers:
    ax.bar_label(i)

# Adding title and labels

plt.title('Lunch Distribution In Data')
plt.xlabel('Lunch')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.show()
```

```python
value_counts = data["Section"].value_counts()
plt.figure(figsize=(12, 6)) 
ax = value_counts.plot(kind='barh', color="lightblue", edgecolor='black')

# Adding count at the top of each bar
for i in ax.containers:
    ax.bar_label(i)

# Adding title and labels

plt.title('Lunch Distribution In Data')
plt.xlabel('Lunch')
plt.ylabel('Count')
plt.xlim(0,11000)
plt.xticks(rotation=0)

plt.show()
```
